```python
from pyspark.sql import SparkSession
```

```python
# Initialize SparkSession with master URL and HDFS configuration
spark = SparkSession.builder \
    .appName("UploadFileToHDFS") \
    .master("spark://spark-master-service:7077") \
    .config("spark.hadoop.fs.defaultFS", "hdfs://172.18.0.2:9000") \
    .getOrCreate()
```

```python
# Replace '/path/to/titanic.csv' with the actual path inside the Spark container
# df = spark.read.csv("file:///workspace/titanic/titanic.csv", header=True, inferSchema=True)
df = spark.read.csv("file:///workspace/titanic/titanic.csv", header=True, inferSchema=True)
```

```python
# Proceed with your analysis
df.show(5)
```

```python
# Write DataFrame to HDFS
df.write \
    .mode('overwrite') \
    .option("header", "true") \
    .csv("/user/root/output/titanic_data")
```
